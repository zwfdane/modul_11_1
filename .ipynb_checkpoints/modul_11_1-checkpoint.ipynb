# Zestawienie cen miedzi oraz wyceny spółki KGHM


Celem raportu jest przedstawienie porównanie cen miedzi oraz cen akcji spółki KGHM w latach 2015-2020.
Rozpoczniemy od wczytania odpowiednich bibliotek oraz dostępnych danych.

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)

import cufflinks as cf
cf.go_offline()

import plotly.offline as pyo
import plotly.graph_objs as go
import chart_studio.plotly as py

from plotly.subplots import make_subplots


import pandas as pd
df_kghm = pd.read_csv('kgh_d.csv')
df_cu = pd.read_csv('ca_c_f_d.csv')
```

```python
# Tworzymy obiekt figure
fig = go.Figure()
# Tworzymy podwykresy dla cen akcji KGHM i cen miedzi ze wspólną osią
fig = make_subplots(rows=2, cols=1,shared_xaxes=True, subplot_titles= ('KGHM','Miedź'))
# Tworzymy podwykres dotyczący cen akcji KGHM
fig.add_trace(go.Scatter(x=df_kghm['Data'],y=df_kghm['Zamkniecie']),1,1)
# Tworzymy podwykres dotyczący cen miedzi
fig.add_trace(go.Scatter(x=df_cu['Data'],y=df_cu['Zamkniecie']),2,1)
# Usuwamy legendę
fig.update_layout(showlegend=False)
fig.show()
```

```python
# Łączymy dane dotyczące KGHM i miedzi
df = pd.merge(df_kghm, df_cu, on = 'Data', how = 'inner')
# Zmieniamy nazwy kolumn
df.rename(columns = {'Zamkniecie_x':'KGHM', 'Zamkniecie_y':'Miedź'}, inplace = True)
# Wybieramy kolumny do analizy
df = df[['Data','KGHM','Miedź']]
```

```python
# Tworzymy wykres składający się z dwóch podwykresów oraz tabeli
fig = make_subplots(
    rows=3, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.03,
    specs=[[{"type": "table"}],
           [{"type": "scatter"}],
           [{"type": "scatter"}]],
    subplot_titles= ('Tabela cen', 'KGHM', 'Miedź')
)
# Tworzymy podwykres dotyczący cen akcji KGHM  - ustawiamy wykres w drugim wierszu
fig.add_trace(
    go.Scatter(
        x=df["Data"],
        y=df["KGHM"],
        mode="lines",
        name="KGHM"
    ),
    row=2, col=1
)
# Tworzymy podwykres dotyczący cen miedzi - ustawiamy wykres w drugim wierszu
fig.add_trace(
    go.Scatter(
        x=df["Data"],
        y=df["Miedź"],
        mode="lines",
        name="Miedź"
    ),
    row=3, col=1
)
# Tworzymy tabelę z wykorzystaniem danych
fig.add_trace(
    go.Table(
        header=dict(
            values=["Data", "KGHM", "Miedź"],
            font=dict(size=10),
            align="left"
        ),
        cells=dict(
            values=[df[k].tolist() for k in df.columns[0:]],
            align = "left")
    ),
    row=1, col=1
)
fig.update_layout(
    showlegend=False,
    title_text="Ceny miedzi oraz akcji KGHM w latach 2015-2020",
)

fig.show()
```

Wniosek. Wykresy szeregów czasowych cen miedzi oraz cen akcji KGHM wykazują duże podobieństwo.
